## 3. Target Distribution Analysis (Log Transform)
Due to right skewness, we analyze both raw and log-transformed `SalePrice`.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df[target_col], kde=True, ax=axes[0], color='royalblue')
axes[0].set_title("Original SalePrice Distribution")

sns.histplot(np.log1p(df[target_col]), kde=True, ax=axes[1], color='teal')
axes[1].set_title("Log-Transformed SalePrice Distribution")
plt.show()
```
